```python
'''
Problem 4:                                                                                                                    
Design a hotel room managment system

Requirement:
1. I should be able to add hotel info i.e name and address
2. I should be able to add new room info i.e name, num_of_beds, fare_per_day                                                 
3. I should be able to book a room for a user/person/account for today and future dates (ignore timezone)
4. I should be able to view the few stats i.e total rooms, total room occupied today, total rooms available today
5. I should be able to check if room is available on future date

Design a good model. Its a simple requirment there will no need to use inheritance
'''
from datetime import date , datetime

class Hotel:
    def __init__(self,hotel_name,hotel_address) -> None:
        self.hotel_name = hotel_name
        self.hotel_address = hotel_address
        self.rooms = []
    
    def add_room_info (self, name,num_of_beds , fare_per_day):
        assert num_of_beds > 0,'Please enter positive number'
        assert fare_per_day > 0 , 'Please enter valid fare'
        
        self.rooms.append({
            "name":name,
            "num_of_beds":num_of_beds,
            "fare_per_day":fare_per_day,
            "id" : len(self.rooms) + 1,
            "bookings" : []
        })
        
    def book_room(self  , start_date , end_date , room_id, user_name ):
        user_start_date = datetime.fromisoformat(start_date).timestamp()
        user_end_date = datetime.fromisoformat(end_date).timestamp()
        # checking wheather room is exists or not 
        if room_id > 0 and room_id <= len(self.rooms):
            for index,room in enumerate(self.rooms):
                # filterted selected room 
                if room['id']==room_id:
                    # if this room is not book yet then add new booking 
                    if len(room['bookings']) == 0:
                        room['bookings'].append({
                            "start_date" : start_date,
                            "end_date": end_date,
                            "isBook":True,
                            "user_name":user_name
                        })
                    else:  
                        isBooked = False  
                        # if this room is already book then checking if my days are available or not 
                        for booking in room['bookings']:
                            # print('bookings',booking,start_date,user_name)
                            book_start_date =datetime.fromisoformat(booking['start_date']).timestamp()
                            book_end_date =datetime.fromisoformat(booking['end_date']).timestamp()
                            if book_start_date <= user_start_date <= book_end_date or book_start_date <= user_end_date <= book_end_date:
                                isBooked = True
                                break 
                        if isBooked == True:
                                print('Room is not available on that days')
                                   
                        else:
                            room['bookings'].append({
                                "start_date" : start_date,
                                "end_date": end_date,
                                "isBook":True,
                                "user_name":user_name
                            })
                    break
        else:
            print('This room is not exist')
    
    def check_stats (self): 
        current_date = date.today()
        total_book_today = 0
        total_available_today =0
        for room in self.rooms:
            # if there is noy booking for this room then it will be available 
            if len(room['bookings']) == 0:
                total_available_today +=1
                
            # otherwise else will call     
            else:
                isBook=False
                for booking in room['bookings']:
                    book_start_date =date.fromisoformat(booking['start_date'])
                    book_end_date =date.fromisoformat(booking['end_date'])
                    if book_start_date <= current_date <= book_end_date:
                       isBook = True
                       break
                if isBook :
                       total_book_today +=1
                else:
                    total_available_today +=1
        stats ={
            "total_rooms": len(self.rooms),
            "total_book_today":total_book_today,
            "total_available_today":total_available_today
        } 
        return stats
     
    def available_rooms(self,_date):
        available_rooms = []
        for room in self.rooms:
            # if there is noy booking for this room then it will be available 
            if len(room['bookings']) == 0:
                available_rooms.append(room)
                
            # otherwise else will call     
            else:
                isBook=False
                for booking in room['bookings']:
                    book_start_date =date.fromisoformat(booking['start_date'])
                    book_end_date =date.fromisoformat(booking['end_date'])
                    if book_start_date <= date.fromisoformat(_date) <= book_end_date:
                       isBook = True
                       break
                if isBook :
                       print('')
                else:
                    available_rooms.append(room)
        
        return available_rooms       
      
hotel = Hotel("Ramada",'abc address')
hotel.add_room_info('average' , 2 , 5000)          
hotel.add_room_info('2 star' , 1 , 3000)          
hotel.add_room_info('5 star' , 3 , 10000)          
hotel.add_room_info('7 star' , 4 , 20000)          
# display(hotel.rooms)
hotel.book_room('2024-07-26','2024-07-26',1,'waqas')
hotel.book_room('2024-07-27','2024-07-27',1,'ali')
hotel.book_room('2024-07-28','2024-07-29',1,'rohan')
hotel.book_room('2024-07-27','2024-07-28',2,'haris')
# hotel.check_stats()
# display(hotel.rooms)
display(hotel.check_stats())
display(hotel.available_rooms('2024-07-31'))
```
